Define transaction categories
---
After the data is loaded to database. I started label transactions with their corresbonding transactions. To do this, I will need to have a list of categories for my transactions. I decided to used below categrizes.



| Category        | Examples |
|:-------------   |:-------------|
|Food and Drink   |Restaurant, Food Hall, Coffee, Bar|
|Transfer         |Bank Transfer, ATM, Vemo,Chase quickpay|
|Shops            |Department Store, Cloth, Electronics, Amazon|
|Service          |Rental, Utilities, Mortgage, Auto Insurance|
|Payroll          |Pay check|
|Payment          |Credit Card payments and others payments|
|Education        |Book, Online courses, Subscriptions|
|Bank Fees        |Membership fee for credit card|
|Grocery          |Vegetable, Fruit and many others|
|Interest         |Interest payment from deposit|
|Travel           |Gas Station, Toll fees, Parking, Flight, Bus, Train, Hotel|
|Recreation       |Tours, Museums, Movies|
|Healthcare       |Dentist visit, Glasses, Pharmacy Stores|
|Tax              |Tax return|


While I was thinking about what categories should I use, I realized that there are numerious ways to do this. **It really depends on individual's preference that how one want to track their transaction.** For example, I would like to have a category (which I named **Service**) that summaries all my fixed or mandatory spendings, like rent and mortgage. I also put transaction like haircut, stamps into this category, because they are mandatory expenses. For these transactions, I put them under 'Service' category. I also made a category 'Payment' to track all my credit card payment, which I want to keep an eye on. I tried not to use too many categories because I don't have large amount of transactions and also I believe these categories are alrealy a good summary of my spendings. Check out a more comprehensive list at  Yodlee's website [Yodlee's transaction categorization](https://developer.yodlee.com/Yodlee_API/Transaction_Categorization)




Explore the data
---

```python
from pymongo import MongoClient
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**Numbers of transactions in each category:** let's see what categories we have in this dataset. It would be nice to see how transactions are distributed accross the categories.

```python
mongoclient = MongoClient('localhost')
##Mongoclient.database_names()
db = mongoclient['bankdata']

summary = pd.DataFrame(list( db.transactions.aggregate([
    {"$group" : {"_id":"$category", "count":{"$sum":1}}}
])))
summary.rename(columns={'_id': 'category'}, inplace=True)
summary.sort_values(by="count", inplace=True)
print(summary.sort_values(by="count"))
#for document in cursor:
#   print(document["_id"]+": "+str(document["count"]))
```

```python
objects = summary["category"]
y_pos = np.arange(len(objects))
performance = summary["count"]

plt.figure(figsize=(8,5)) 
plt.bar(y_pos, performance, align='center', alpha=0.8)
plt.xticks(y_pos, objects,fontsize=10, rotation=90)
plt.ylabel('Number of Transactions')
plt.title('Number of transactions in each category', fontsize=12)
plt.show()
```

Food and Drink appreas to me the most popular transactions with almost 400 transactions. Transfer, shops and payments each has around 200 transactions. Payroll, Grocery, Travel, and Server are in the 100 to 150 range.
It can be seen that there is only 1 transaction in category **Tax** and 3 transactions in **Bank Fees**.
